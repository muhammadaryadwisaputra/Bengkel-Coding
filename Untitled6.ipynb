```python
# 2. Import library
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# 3. Load dataset (ganti path ini dengan yang benar sesuai lokasi file di Drive kamu)
df = pd.read_csv('/content/drive/MyDrive/DataBK/ObesityDataSet.csv')

# 4 Tampilkan beberapa baris pertama
print("5 Baris Pertama Dataset:")
print(df.head())

# 5. Informasi Umum Dataset
print("\nInformasi Dataset:")
print(df.info())

print("\nDeskripsi Statistik:")
print(df.describe(include='all'))

# 6. Cek Missing Values
print("\nJumlah Missing Values per Kolom:")
print(df.isnull().sum())

# 7. Cek Unique Values
print("\nJumlah Nilai Unik per Kolom:")
print(df.nunique())

# 8. Cek Duplikasi
print("\nJumlah Data Duplikat:")
print(df.duplicated().sum())

# 9. Visualisasi Distribusi Target
plt.figure(figsize=(10,5))
sns.countplot(data=df, x='NObeyesdad', order=df['NObeyesdad'].value_counts().index)
plt.title("Distribusi Target (Obesity Level)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# 10. Boxplot untuk Deteksi Outlier (variabel numerik)
numerical_features = ['Age']
plt.figure(figsize=(14, 10))  # lebih besar agar tidak saling tumpuk

for i, feature in enumerate(numerical_features):
    plt.subplot(3, 3, i + 1)
    sns.boxplot(data=df, x=feature, color='skyblue')
    plt.title(f'Boxplot of {feature}', fontsize=12)
    plt.xlabel("")  # Hilangkan label agar tidak menumpuk
    plt.tight_layout()

plt.suptitle('Boxplot untuk Deteksi Outlier', fontsize=18, y=1.02)
plt.show()
```
